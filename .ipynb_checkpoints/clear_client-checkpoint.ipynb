## Очистка колонок, название колонок, удаление лишних колонок, удаление первых строк, удаление полностью пустых строк

```python
df_client = df_client.iloc[:, 0:14]
df_client.columns = ['id', 'name_user', 'phone', 'city', 'addres', 'kids', 'status', 'source', 
                          'call_23_24', 'estimation', 'note_23_24', 'skip', 'old_call', 'old_note']
df_client = df_client.drop(['estimation', 'skip', 'old_call', 'old_note'], axis=1)
df_client = df_client[4:]
df_client = df_client.dropna(how = 'all')
```

### Города без повторений

```python
df_client['city'] = df_client['city'].replace(['Старая Купавна ', 'СтараяКупавна', 'Старая купавна '],'Старая Купавна')
df_client['city'] = df_client['city'].replace(['Новая\nКупавна','Новая купавна'], 'Новая Купавна')
df_client['city'] = df_client['city'].replace(['г. Балашиха', 'Балашиха '],'Балашиха')
df_client['city'] = df_client['city'].replace('монино','Монино')
df_client['city'] = df_client['city'].replace('Элекктросталь','Электросталь')
df_client['city'] = df_client['city'].replace(' Кучино','Кучино')
df_client['city'] = df_client['city'].replace(['Железнодорожный ', 'Железнодорожная '],'Железнодорожный')
df_client['city'] = df_client['city'].replace('Бисерово ','Бисерово')
df_client['city'] = df_client['city'].replace('Обухово ','Обухово')
df_client.city.value_counts()
```

### Проверка статуса клиента

### Изменение источника

```python
df_client['source'] = df_client['source'].replace([' интернет', 'интернет','Интерент'],'Интернет')
df_client['source'] = df_client['source'].replace(['Группа ВК','вконтакте','В контакте','Вконтакте, заказала родитель','Контакт', 'вк', 'ВК'], 'Вконтакте')
df_client['source'] = df_client['source'].replace(['в сад приходил Олег','Знакомые Лизы','знакомые','Лена','Лиза', 'Дима', 'Тамара'],'Знакомые')
df_client['source'] = df_client['source'].replace(['Инстаграм','инстаграм','инстаграмм'],'Инстаграмм')
df_client['source'] = df_client['source'].replace(['Рекомендации','рекомндация','Подруга Алена5','Сосед Екатерина22','рекоминдация'],'Рекомендация')
df_client['source'] = df_client['source'].replace('авито','Авито')
df_client['source'] = df_client['source'].replace(['Выиграла в конкурсе','Конкурс вк', '1место '],'Конкурс')
df_client['source'] = df_client['source'].replace('Визитки','Визитка')
df_client['source'] = df_client['source'].replace('Авто','Машина')
df_client.source.value_counts()
```

```python
df_client['call_23_24'] = df_client['call_23_24'].replace(['уежаю ', 'уезжают','уехали','бдут праздновать в дурго месте','уезжает','Уезжаюь','Уезжайт'],'Уезжают')
df_client['call_23_24'] = df_client['call_23_24'].replace(['Заказала','заказала','заказал','закала','Заказал ','Заказала ','Зкаазала'], 'Заказал')
df_client['call_23_24'] = df_client['call_23_24'].replace(['занято','не ответил','не отвечает','не доступен','не ответила','анято','автоотвечик','сбросила','Занята','скинула'],'Занято')
df_client['call_23_24'] = df_client['call_23_24'].replace(['думают','уточнит','думает','Думает','пока не думали','Думает ','Напишет','Если что перезвонит','Перезвонит','Ответит 14 числа','Отпишется если что ','Скинул цены','Пока не знаем ','перезвонит'],'Думают')
df_client['call_23_24'] = df_client['call_23_24'].replace(['заказали у других','ушли к конкурентам','Сосед Екатерина22','рекоминдация'],'Ушли к конкурентам')
df_client['call_23_24'] = df_client['call_23_24'].replace(['номер не але','номер не обслуживается','Абонент в сети не зарегистрирован'],'Номер заблокирован')
df_client['call_23_24'] = df_client['call_23_24'].replace(['Отменила','Передумала ','отказалась'],'Отменил')
df_client['call_23_24'] = df_client['call_23_24'].replace(['переехали','Переехали в Таганрог','Переехали в Москву','Перехали в Москву'],'Переехали')
df_client['call_23_24'] = df_client['call_23_24'].replace(['дети не хотят','Дети не хотят','выросли','вырос','выросли ','Вырочли дети','Вырос ребенок', 'выросли дети', 'Не будут из-за ковида и выросли дети', 'Дети выросли', 'Вырасли дети'],'Выросли дети')
df_client['call_23_24'] = df_client['call_23_24'].replace(['не планируют','не будут','не будут в этом году','не получается','пригласили на елки','финансы','сегодня','на', '!','не будут заказывать','-','не будут ','не ответили','написал', 'написал', 'не будут ', 'сама играет снегурочку','Не Заказал ','Не ответила ', 'Не отвечает ','Не ответил', 'Уже провели ярмарку', 'Не ответила',
       'не стал звонить','Не отвечает','Не интересует','Не нужно', 'не решились'],'Не будут(в этом году)')
df_client['call_23_24'] = df_client['call_23_24'].replace('заболели','Заболели')
df_client.call_23_24.unique()
```

## Разделение колонки с телефона и чистка под единоую форму 

```python
df_client['phone']
```

```python
df_client['phone'] = df_client['phone'].astype(str)
df_client['phone'] = df_client['phone'].str.replace('7-903-1911-87-29','89031918729')
df_client['phone'] = df_client['phone'].str.replace('7-965-42-77-04','89654207704')
df_client['phone'] = df_client['phone'].str.replace('-','')
df_client['phone'] = df_client['phone'].str.replace('!','')
df_client['phone'] = df_client['phone'].str.replace('(','')
df_client['phone'] = df_client['phone'].str.replace(')','')
df_client['phone'] = df_client['phone'].str.replace(',','')
df_client['phone'] = df_client['phone'].str.replace('--','')
df_client['phone'] = df_client['phone'].str.replace('дом','')
df_client['phone'] = df_client['phone'].str.replace('Камил','')
df_client['phone'] = df_client['phone'].str.replace('Андрей','')
df_client['phone'] = df_client['phone'].str.replace('Радмила','')
df_client['phone'] = df_client['phone'].str.replace(';','')
df_client['phone'] = df_client['phone'].str.replace('+','')
df_client['phone'] = df_client['phone'].str.replace('-','')
df_client['phone'] = df_client['phone'].str.replace(' ','')
df_client['phone'] = df_client['phone'].str.replace('  ','')
df_client['phone'] = df_client['phone'].str.replace('\n','')
df_client['phone'] = df_client['phone'].str.replace("'",'')
df_client['phone'] = df_client['phone'].str.replace('\u202a','')
df_client['phone'] = df_client['phone'].str.replace('\xa0','')
df_client['phone'] = df_client['phone'].str.replace('\u202c','')
df_client['phone'] = df_client['phone'].str.replace('‑','')
```

```python
df_client['phone']
```

### Разделение номеров

```python
phone = []
for row in df_client['phone']:
    if len(row) == 11:
        row = row + ',0,0'
    elif len(row) == 22:
        row = row[:11] + ',' + row[11:] + ',0'
    else:
        row = row[:11] + ',' + row[11:22] + ',' + row[22:]
    phone.append(row)
df_client['phone'] = phone
```

```python
df_client[['phone_1','phone_2']] = df_client['phone'].str.split(',',1,expand=True)
df_client[['phone_2','phone_3']] = df_client['phone_2'].str.split(',',1,expand=True)
```

```python
df_client
```

### Изменение первого символа в номере

```python
phone = []
for row in df_client['phone_1']:
    if row != '0':
        final_str = '8' + row[1:]
        phone.append(final_str)
    else:
        phone.append('0')
df_client['phone_1'] = phone
```

```python
phone = []
for row in df_client['phone_2']:
    if row != '0':
        final_str = '8' + row[1:]
        phone.append(final_str)
    else:
        phone.append('0') 
df_client['phone_2'] = phone
```

```python
phone = []
for row in df_client['phone_3']:
    if row != '0':
        final_str = '8' + row[1:]
        phone.append(final_str)
    else:
        phone.append('0') 
df_client['phone_3'] = phone
```
