## Создание excel файла за 16-17 год из файла 17-18 года

#### Копирование файла

```python
df_client_16_17 = df_client_17_18.copy(deep=True)
```

#### Название колонок

```python
df_client_16_17.columns = ['id', 'name_user', 'phone', 'city', 'addres', 'kids', 'status', 'source', 
                          'new_call', 'estimation', 'new_note', 'skip', 'call', 'note']
```

#### Удаление колонок нужных для 17-18 года

```python
df_client_16_17 = df_client_16_17.drop(['new_call', 'estimation', 'new_note', 'skip'], axis=1)
```

#### Удаление первый пустых строк

```python
df_client_16_17 = df_client_16_17[4:]
```

#### Удаление пустых строк

```python
df_client_16_17 = df_client_16_17.dropna(how = 'all')
```

#### Выбираем строки по "звонку" и "примечание" когда заполнялись в 16-17 году

```python
df_client_16_17 = df_client_16_17[:282]
```
